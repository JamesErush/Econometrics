```python
standard_deviationln = stat.stdev(ln_ahe)
```

```python
outlier_test = 3*standard_deviationln
```

```python
lower_bound = ln_ahe_mean - outlier_test
```

```python
ln_ahe_list = [2.5708491020281183]
```

```python
for x in ln_ahe:
    ln_ahe_list.append(x)
```

```python
sci.stats.describe(ln_ahe_list)
```

```python
for x in ln_ahe_list:
    yrseduc_index = ln_yrseduc_list.index(x)
    if lower_bound > x:
        ln_yrseduc_list.remove(ln_yrseduc_list[yrseduc_index])
        ln_ahe_list.remove(ln_ahe_list[yrseduc_index])
    elif x > upper_bound:
        ln_yrseduc_list.remove(ln_yrseduc_list[yrseduc_index])
        ln_ahe_list.remove(ln_ahe_list[yrseduc_index])
```
